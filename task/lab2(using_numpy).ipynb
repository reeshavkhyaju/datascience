a. Addition
b. Subtraction

```python
import numpy as np 

#init a matirx
a = np.array([ [13, 14], [17, 19] ])
a
```

4. Collect the age information from the lab groupmates via. google form then, compute
mean, median, mode, and standard deviation.      Use pandas to read the Excel file and
numpy for further calculation. Normalize the age of your groupmates using the
following formula:    
X = (X - Xmean) / Xstd

```python
import pandas as pd

df = pd.read_csv("age.csv")

df.columns
```

```python
df = df.rename(columns = {"How long have you existed?": "age"})
df["age"] = df["age"].astype(int)

ageMean = df["age"].mean()
ageSD = df["age"].std()

df["normalized_age"] = (df["age"] - ageMean) / ageSD

for age in df["normalized_age"]:
    print(age)
```

5. Contemplate the difference between scaling and normalization techniques that are
used in tasks 3 and 4

>ANSWER:<br><br>
In task 3, the operation <b>( X = (X - Xmin) / (Xmax - Xmin) )</b> scales the data to a range of [0, 1]. This operation makes the smallest value 0 and the largest value 1. Scaling in a range specific transformation ensures the data fits that specified rangee. This is commonly used for machine learning algorithms sensitive to range of input. The relative relationships between the data points is preserved.<br>
In task 4, the operation <b>( X = (X - Xmean) / Xstd )</b> standardizes the data by centering it around a mean of 0 and scaling it based on the standard deviation. This operation produces data with a mean of 0 and a standard deviation of 1. This is commonly used when data follows or approximately follows a normal distribuition.<br>
<br>While scaling is more about transforming the data into a defined range for special use in magnitude sensitive algorithms, like the gradient descent.<br>
Normalization standardizes the data such that features with different units or distribuitions can be comparable. For example, in medical records, non-normalized data might value features with higher numeric ranges like weight over others, potentially skewing analysis/predictions.
