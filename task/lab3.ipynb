```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, binom, poisson, norm
```

a. Bernoulli

```python
#parameters
p_bernoulli = 0.5
size_bernoulli = 100

bernoulli_data = bernoulli.rvs(p=p_bernoulli, size=size_bernoulli)
sns.histplot(bernoulli_data, bins=2, shrink=0.8, color='red', alpha=0.7)

# graph
plt.title(f'Bernoulli Distribution of p={p_bernoulli}')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()

# mean and median of bernoulli data
mean_bernoulli = np.mean(bernoulli_data)
median_bernoulli = np.median(bernoulli_data)
print(f"Mean of Bernoulli data: {mean_bernoulli}")
print(f"Median of Bernoulli data: {median_bernoulli}")
```

b. Binomial

```python
# Parameters for the Binomial Distribution
n_binomial = 10  # Number of trials
p_binomial = 0.5  # Probability of success in each trial
size_binomial = 1000  # Number of experiments

# Generate random data based on the binomial distribution
binomial_data = binom.rvs(n=n_binomial, p=p_binomial, size=size_binomial)

# Plot the histogram
plt.hist(binomial_data, bins=n_binomial + 1, rwidth=0.8, color='lightgreen', alpha=0.7)
plt.title('Binomial Distribution')
plt.xlabel('Number of Successes (Heads)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)  # Add grid for better readability
plt.show()

# mean and median of binomial data
mean_binomial = np.mean(binomial_data)
median_binomial = np.median(binomial_data)
print(f"Mean of Binomial data: {mean_binomial}")
print(f"Median of Binomial data: {median_binomial}")
```

c. Poisson Distribution

```python
# Poisson Distribution
#parameters
mu_poisson = 3
size_poisson = 1000

poisson_data = poisson.rvs(mu=mu_poisson, size=size_poisson)
plt.hist(poisson_data, bins=np.max(poisson_data)-np.min(poisson_data)+1, rwidth=0.8, color='salmon', alpha=0.7)
plt.title('Poisson Distribution')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()

# Calculate and print mean and median of the Poisson data
mean_poisson = np.mean(poisson_data)
median_poisson = np.median(poisson_data)
print(f"Mean of Poisson data: {mean_poisson}")
print(f"Median of Poisson data: {median_poisson}")
```

d. Gaussian Distribution

```python
# Gaussian Distribution
#parameters
mean_gaussian = 0
std_gaussian = 1
size_gaussian = 1000

gaussian_data = norm.rvs(loc=mean_gaussian, scale=std_gaussian, size=size_gaussian)
plt.subplot(2,1,2)
plt.hist(gaussian_data, bins=30, rwidth=0.8, color='plum', alpha=0.7)
plt.title('Gaussian Distribution')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()

# Calculate and print mean and median of the gaussian data
mean = np.mean(gaussian_data)
median = np.median(gaussian_data)
print(f"Mean of Gaussian data: {mean}")
print(f"Median of Gaussian data: {median}")
```

## 2. Observe and note the effect of parameters on the above distributions.

# a. Bernoulli Distribution

Parameter: p_bernoulli 
When p_bernoulli is closer to 0, most values will be 0 (failure).

When p_bernoulli is closer to 1, most values will be 1 (success).

A balanced p_bernoulli = 0.5 leads to nearly equal occurrences of 0s and 1s.

# b. Binomial Distribution

Parameters: Number of trials (n_binomial), Probability of success in each trial(p_binomial),Number of experiments Number(size_binomial)
Increasing n spreads the distribution, making it cover a larger range of successes.

A higher p_binomial skews the distribution toward higher numbers of successes, while a lower p_binomial skews it toward fewer successes.

When p_binomial = 0.5 and n is larger, the distribution becomes more symmetric.

# c. Poisson Distribution

Parameter: Mean Rate (mu_poisson)

Increasing mu_poisson shifts the distribution center to the right, resulting in more frequent occurrences of larger values.
Lower mu_poisson values concentrate the distribution closer to 0, making it peakier.
For large mu_poisson, the distribution starts to resemble a Gaussian (normal) distribution.

# d. Gaussian Distribution

Parameters: Mean (mean_gaussian) and Standard Deviation (std_gaussian)

Changing the mean shifts the entire distribution along the x-axis without altering its shape.
Increasing std spreads the distribution wider, reducing the height of the peak, while a smaller std makes it narrower and taller.



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv("age_data.csv")
# Extract the 'Age' column
ages = data['Age']

# Calculate mean and standard deviation
mean = ages.mean()
std_dev = ages.std()

# Generate Gaussian distribution data points
x = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 1000)
y = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)

# Plot the graph
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f"Mean = {mean:.2f}, Std Dev = {std_dev:.2f}")
plt.title("Gaussian Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Probability Density")
plt.hist(ages, bins=10, density=True, alpha=0.5, label="Age Data Histogram")
plt.legend()
plt.grid()
plt.show()
```
